```python
import requests
from tqdm import tqdm
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64; rv:45.0) Gecko/20100101 Firefox/45.0'}
url = 'https://lifehacker.ru/topics/technology/'
response = requests.get(url, headers = headers)
soup = BeautifulSoup(response.text, 'lxml')
```

```python
links = []
for i in tqdm(range(10)):
    url = f'https://lifehacker.ru/topics/technology/?page={i + 1}'
    response = requests.get(url, headers = headers)
    soup = BeautifulSoup(response.text, 'lxml')
    
    #raw_items = soup.find_all('a', class_='lh-small-article-card__link') 
    # таким методом в список попадала статья-реклама https://lifehacker.ru/special/vtb-trendy/
    # с отличающейся разметкой страницы, я решила её обходить, добавив ещё один идентификатор 
    
    raw_items = soup.find_all('a', class_='lh-small-article-card__link', target='')
    links += [item.get('href') for item in raw_items]
    time.sleep(1)
```

```python
links[:5]
```

```python
result = []
for ref in tqdm(links):
    article = {}
    content = []
    url = 'https://lifehacker.ru/' + ref
    response = requests.get(url, headers = headers)
    soup = BeautifulSoup(response.text, 'lxml')
    title = soup.find('h1', class_='article-card__title').text
    row_content = soup.find_all('p', class_='')
    content += [content.text for content in row_content]
    article['title'] = title
    article['text'] = " ".join(content)
    result.append(article)
    time.sleep(1)
```

```python
pd.set_option('display.max_colwidth', 400)
df = pd.DataFrame(result)
df.head()
```
