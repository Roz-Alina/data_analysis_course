```python
df_house = pd.read_csv("Housing.csv")
df_house.head(3)
```

```python
df_house.info()
```

```python
df_house.describe()
# Для столбцов bedrooms, bathrooms, stories и parking оптимальным типом данных является np.int8
# Для столбца price -- np.int32
# Для столбца area -- np.int16
```

```python
df_house["price"] = df_house["price"].astype(np.int32)
df_house["area"] = df_house["area"].astype(np.int16)
df_house["bedrooms"] = df_house["bedrooms"].astype(np.int8)
df_house["bathrooms"] = df_house["bathrooms"].astype(np.int8)
df_house["stories"] = df_house["stories"].astype(np.int8)
df_house["parking"] = df_house["parking"].astype(np.int8)
df_house.info()
# Уменьшили вес датасета на 20 КБ
```
