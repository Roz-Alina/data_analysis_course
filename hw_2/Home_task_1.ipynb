```python
class Chocolate:
  def __init__(self, price, inclusion = None):
    self.price = price
    self.inclusion = inclusion

    if self.price < 0:
      raise ValueError("Цена не может быть отрицательной")

  def show_my_chocolate(self):
    if self.inclusion == None:
      print('Обычный шоколадный батончик')
    else:
      print(f'Шоколадный батончик, добавка: {self.inclusion}')

  def __str__(self):
    return(f'Шоколадный батончик, {self.price} рублей')
```

```python
kinder = Chocolate(10.5)
snickers = Chocolate (15.7, 'орехи')
```

```python
kinder.show_my_chocolate()
```

```python
snickers.show_my_chocolate()
```

```python
print(kinder)
print(snickers)
```
